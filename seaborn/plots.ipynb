Count Plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example dataset with different names
data = {'Name': ['Kasun', 'Anuradha', 'Hasitha', 'Nuwan', 'Sarana', 
                    'Madura', 'Vinodh', 'Sandun', 'Kavinda', 'Nadun'
                 'Malith', 'Dimuth', 'Shehan', 'Tharindu', 'Nimesh']}

df = pd.DataFrame(data)

# Setting the color palette
sns.set_palette("pastel")

# Creating the count plot
plt.figure(figsize=(10, 6))  # Setting the figure size
ax = sns.countplot(x='Name', data=df)

# Adding count labels on top of each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='baseline', color='black', fontsize=12, 
                xytext=(0, 5), textcoords='offset points')

# Display the plot
plt.title('Count of Names', fontsize=16)
plt.xlabel('Name', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.grid(axis='y')  # Adding grid lines for better readability
plt.show()
```

KDE Plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example dataset
data = {'Age': [30, 21, 29, 28, 25, 35, 24, 31, 27, 22, 
                30, 26, 22, 23, 29, 32, 34, 28, 27, 25]}

df = pd.DataFrame(data)

# Creating the KDE plot
plt.figure(figsize=(10, 6))  # Setting the figure size
sns.kdeplot(df['Age'], fill=True, color='blue', alpha=0.5)

# Display the plot
plt.title('KDE Plot of Age Distribution', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.grid(True)  # Adding grid lines for better readability
plt.show()
```

Customizing Seaborn Plots 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'Name': ['Kasun', 'Anuradha', 'Hasitha', 'Nuwan', 'Sarana', 
             'Malith', 'Dimuth', 'Shehan', 'Tharindu', 'Nimesh'],
    'Age': [30, 21, 29, 28, 25, 35, 24, 31, 27, 22],
    'Weight': [70, 65, 75, 68, 62, 80, 64, 78, 66, 60]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Change the plot style
sns.set_style("whitegrid")  # Options: 'white', 'dark', 'whitegrid', 'darkgrid', 'ticks'

# Customize the color palette
sns.set_palette("husl")  # Options: 'deep', 'muted', 'bright', 'pastel', 'dark', 'colorblind', or use custom colors

# Set the figure size and aspect ratio
plt.figure(figsize=(20, 6))  # Width, Height in inches

# Creating a bar plot
sns.barplot(x='Name', y='Weight', data=df)

# Adding title and labels
plt.title('Weight of Individuals', fontsize=18)  # Title of the plot
plt.xlabel('Names', fontsize=14)                  # X-axis label
plt.ylabel('Weight (kg)', fontsize=14)            # Y-axis label

# Adjusting the aspect ratio (optional, can also be done via figsize)
plt.gca().set_aspect(0.1)  # Adjusts the aspect ratio

# Show gridlines for better readability
plt.grid(axis='y')

# Display the plot
plt.show()
```
