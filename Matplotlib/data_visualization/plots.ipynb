```python
import matplotlib.pyplot as plt

X = [10,20,30,40,50]
Y = [20,28,36,44,54]

plt.plot(X,Y)

plt.title("Line Chart")

plt.ylabel('Y-axis')
plt.xlabel('X-axis')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('covid-data.csv')
# covid-19 data set about number of cases in a day

x = data['date']
y = data['total_cases']

plt.bar(x,y)

plt.title("Bar Chart")

plt.ylabel('Total Cases')
plt.xlabel('Day')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('Household energy bill data.csv')

x = data['housearea']

plt.hist(x)

plt.title('histogram')

plt.ylabel('amount_paid')
plt.xlabel('thousearea')

plt.show()

```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('Household energy bill data.csv')

x = data['num_rooms']
y = data['amount_paid']

plt.scatter(x,y)

plt.title("scatter plot")

plt.ylabel('amount_paid')
plt.xlabel('housearea')

plt.show()
```

```python
import matplotlib.pyplot as plt


cars = ['AUDI', 'BMW', 'FORD', 'TESLA', 'JAGUAR']
data = [23,17,35,29,12]

plt.pie(data, labels=cars)

plt.title('Pie Chart')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(10)
data = [np.random.normal(0,std,100) for std in range(1,4)]

plt.boxplot(data, vert=True, patch_artist=True,boxprops=dict(facecolor='skyblue'),medianprops=dict(color='red'))

plt.xlabel('Data Set')
plt.ylabel('Values')
plt.title('box plot example')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
data = np.random.rand(10,10)

plt.imshow(data, cmap='viridis', interpolation='nearest')
plt.colorbar()

plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Heatmap example')

plt.show()
```
