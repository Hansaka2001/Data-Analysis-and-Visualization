axes() method

```python
# add_axes() method 

# Python program to show pyplot module
import matplotlib.pyplot as plt
from matplotlib.figure import Figure

# initializing the data
x = [10, 20, 30, 40, 50]
y = [20, 25, 35, 55, 80]

# Creating a new figure with width = 5 inches
# and height = 4 inches
fig = plt.figure(figsize =(6, 4))

# Creating first axes for the figure
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# Creating second axes for the figure
ax2 = fig.add_axes([1, 0.1, 0.8, 0.8])

# Adding the data to be plotted
ax1.plot(x, y)
ax2.plot(y, x)

plt.show()
```

Subplot() method

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40, 50]
y = [20, 25, 35, 55, 80]

plt.figure()

plt.subplot(121)
plt.plot(x, y)

plt.subplot(122)
plt.plot(y, x)
```

subplots() method

```python
import matplotlib.pyplot as plt 

x = [10,20,30,40,50]
y = [20,25,35,55,80]

fig, axes = plt.subplots(1,2)

axes[0].plot(x, y)
axes[0].plot(y, x)

axes[1].plot(x, y)
```

subplot2grid() method

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40, 50]
y = [20, 25, 35, 55, 80]

axes1 = plt.subplot2grid(
    (7,1), (0,0), rowspan=2, colspan=1
)

axes2 = plt.subplot2grid(
    (7,1), (2,0), rowspan=2, colspan=1
)

axes1.plot(x, y)
axes2.plot(y, x)
```

Saving the Matplotlib Plots Using savefig()

```python
import matplotlib.pyplot as plt

year = [2010, 2011, 2012, 2013, 2014, 2015, 2016]
production = [100, 120, 240, 160, 140, 200, 170]

plt.bar(year, production)

plt.savefig('barchart_output.jpg')

plt.savefig('output1', facecolor='y', bbox_inches='tight',pad_inches=0.5, transparent=True)
```
