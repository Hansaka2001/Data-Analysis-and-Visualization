```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

plt.plot(x, y)
plt.title('Linear graph', fontsize=25, color='green')

plt.ylabel('y-axis')
plt.xlabel('x-axis')

plt.ylim(0,80)

plt.xticks(x,labels=['one','two','three','four'])

plt.show()
```

```python
import matplotlib.pyplot as plt


x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

# plotting the data
plt.plot(x, y, color='green', linewidth=3, marker='o', 
         markersize=15, linestyle='--')


plt.title("Line Chart with style")
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('covid-data.csv')

x = data['date']
y = data['total_cases']

# plotting the data
plt.bar(x, y, color='green', edgecolor='blue', 
        linewidth=2)

plt.title("covid-19 dataset")

plt.ylabel('date')
plt.xlabel('total cases')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('Household energy bill data.csv')

x = data['housearea']

plt.hist(x, bins=25, color='green', edgecolor='blue', linestyle='--', alpha=0.5)

plt.title('histogram')

plt.ylabel('amount_paid')
plt.xlabel('housearea')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('Household energy bill data.csv')

x = data['num_rooms']
y = data['amount_paid']

plt.scatter(x,y, c=data['housearea'], s=data['amount_paid'], marker='D', alpha=0.5)

plt.title("scatter plot")

plt.ylabel('amount_paid')
plt.xlabel('housearea')

plt.show()
```

```python
import matplotlib.pyplot as plt

cars = ['AUDI', 'BMW', 'FORD', 
        'TESLA', 'JAGUAR']

data = [23, 17, 35, 29, 12]

explode = [0.1,0.5,0.3,0,0]

colors = ( "orange", "cyan" , "yellow", 
          "green", "grey")

plt.pie(data, labels=cars, explode=explode, autopct='%1.2f%%', colors=colors, shadow=True)

plt.show()
```
