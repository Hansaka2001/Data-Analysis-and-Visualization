```python
import matplotlib.pyplot as plt

# initializing the data
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

# plotting the data
plt.plot(x, y)

# Adding title to the plot
plt.title("Linear graph", fontsize=25, color="green")

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

plt.plot(x, y)
plt.title('Linear graph', fontsize=25, color='green')

plt.ylabel('y-axis')
plt.xlabel('x-axis')

plt.ylim(0,80)

plt.xticks(x,labels=['one','two','three','four'])

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=[20,25,35,55]

plt.plot(x,y)
plt.title('Linear graph',fontsize=25,color='green')
plt.ylabel('y-axis')
plt.xlabel('x-axis')
plt.ylim(0,80)
plt.xticks(x,labels=['one','two','three','four'])

# Adding legends
plt.legend(['GFG'])

plt.show()
```

```python
import matplotlib.pyplot as plt


x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

# plotting the data
plt.plot(x, y, color='green', linewidth=3, marker='o', 
         markersize=15, linestyle='--')


plt.title("Line Chart with style")
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')

plt.show()
```
