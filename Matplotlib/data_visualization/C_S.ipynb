```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

plt.plot(x, y)
plt.title('Linear graph', fontsize=25, color='green')

plt.ylabel('y-axis')
plt.xlabel('x-axis')

plt.ylim(0,80)

plt.xticks(x,labels=['one','two','three','four'])

plt.show()
```
