```python
import matplotlib.pyplot as plt

X = [10,20,30,40,50]
Y = [20,28,36,44,54]

plt.plot(X,Y)

plt.title("Line Chart")

plt.ylabel('Y-axis')
plt.xlabel('X-axis')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('covid-data.csv')
# covid-19 data set about number of cases in a day

x = data['date']
y = data['total_cases']

plt.bar(x,y)

plt.title("Bar Chart")

plt.ylabel('Total Cases')
plt.xlabel('Day')

plt.show()
```
